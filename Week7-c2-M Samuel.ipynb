## Exercise 2: Manipulate columns in the dataframe

#### In this exercise we will create new columns, then insert data based on the existing columns.

```python
# Display only the values in one column

education_df['Students Taking AP Exams']
```

#### Review and run the following code, which compares each year's number of AP exam takers to the average across all years

```python
# Use a method that computes the average number of students who took the AP exam, across all years

avg_ap_takers = education_df['Students Taking AP Exams'].mean()
print(avg_ap_takers)
```

```python
# Create a new column that shows the difference between that year's value and the average

education_df['diff_from_avg'] = education_df['Students Taking AP Exams'] - avg_ap_takers
education_df.diff_from_avg
```

#### Now try the same thing for the `stem_college_graduates` column.

```python
def reverse_list(t):
    u = []
    i = len(t) 
    while i > 0:
        u.append(t[i])
        i -= 1
    return u
```
